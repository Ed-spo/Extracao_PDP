```python
# Importando as bibliotecas necessárias
import pandas as pd
import openpyxl
```

# 1. Definir o nome do arquivo, a aba de origem e as unidades de interesse

```python
# 1. Definições
input_file = "/content/PDP 2025 MGI - Necessidades de desenvolvimento_NO_PROCESSO-DEZ_25.xlsm"
sheet_name = "COMPLETA"
unidades_interesse = ['CGBEN', 'CGPAG', 'COADM', 'CGRIS', 'DIGEP', 'COATE']
header_row_index = 5 # Linha 6 do Excel, onde o cabeçalho real foi encontrado
```

```python
arquivo = pd.ExcelFile(input_file)
arquivo.sheet_names
display(arquivo.sheet_names)
```

# 2. Mapeamento FINAL de colunas de ORIGEM (do Excel) para colunas de DESTINO (na Extração)


```python
# Usamos 'Ação de desenvolvimento' como coluna de origem para 'MODALIDADE DE CAPACITAÇÃO'.
column_mapping = {
    'Identificador': 'IDENTIFICADOR',
    'Unidades (IFS 0.1)': 'UNIDADE',
    'Perfil dos agentes públicos: (IFS 0.2)': 'PÚBLICO-ALVO',
    'O que NÃO SE SABE FAZER ou NÃO SER (IFS 1)': 'OBJETIVO DA AÇÃO (IFS 1)',
    'Capacidade a ser desenvolvida (IFS 5)': 'OBJETIVO NA PRÁTICA (IFS 5)',
    'Tema Geral (IFS 3)': 'TEMA GERAL',
    'Recorte do Tema Geral (IFS 3)': 'RECORTE DO TEMA GERAL',
    'Ação de desenvolvimento': 'MODALIDADE DE CAPACITAÇÃO'
}
source_cols = list(column_mapping.keys())
filter_col = 'Unidades (IFS 0.1)'
output_file = "Extracao_Decipex_Resultado_V4.xlsx"
nova_aba = "Extração Decipex"
```

```python
# Solução: Especificar o cabeçalho correto
df_completa = pd.read_excel(
    input_file,
    sheet_name=sheet_name,
    engine='openpyxl',
    header=5  # O cabeçalho real está na 6ª linha do Excel (índice 5)
)
```

# --- Etapas de Processamento ---

```python
try:
    # 3. Ler a aba de origem ("COMPLETA") com a correção do cabeçalho
    df_completa = pd.read_excel(
        input_file,
        sheet_name=sheet_name,
        engine='openpyxl',
        header=header_row_index
    )

    print(f"Dados da aba '{sheet_name}' lidos com sucesso.")

    # 4. Filtrar, 5. Selecionar e Renomear
    df_filtrado = df_completa[df_completa[filter_col].isin(unidades_interesse)]
    df_extracao = df_filtrado[source_cols].copy()
    df_extracao.rename(columns=column_mapping, inplace=True)

    print(f"Extração e renomeação concluída. {len(df_extracao)} registros encontrados.")

    # 6. Salvar os dados extraídos em uma nova aba no arquivo de saída
    with pd.ExcelWriter(output_file, engine='openpyxl') as writer:
        df_extracao.to_excel(writer, sheet_name=nova_aba, index=False)

    # 7. FUNÇÃO DE DOWNLOAD DO COLAB: Baixa o arquivo recém-criado para o seu computador
    files.download(output_file)

    print(f"\n✅ Arquivo '{output_file}' criado com sucesso e download iniciado!")

except FileNotFoundError:
    print(f"❌ ERRO: O arquivo '{input_file}' não foi encontrado. Certifique-se de que ele foi carregado no Colab.")
except KeyError as e:
    print(f"❌ ERRO: Coluna não encontrada: {e}. Verifique se os nomes das colunas de origem estão corretos.")
except Exception as e:
    print(f"❌ Ocorreu um erro inesperado: {e}")
```
